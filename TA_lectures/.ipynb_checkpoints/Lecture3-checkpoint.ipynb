## K-folds Cross Validation


- Divide randomly the observations in k folds 
- Use k-1 folds as a training set
- Make the predictions and compute the MSPE on the remaining fold
- Repeat the process for each fold
- Use the mean of each MSE as an estimator of the MSPE

### Ridge Regression

```r
wine.white <- read.table("./data/wine-white.txt")
# install.packages('glmnet')
library(glmnet)
lambdas.rr <- exp(seq(-4, 4, length = 50))
y <- wine.white$quality

## Construct the covariate matrix
X <- as.matrix(wine.white[,-12])
fit <- glmnet(
  x = X,
  y = y,
  lambda = lambdas.rr,
  alpha = 0
  ) 

# alpha = 1 for lasso, alpha = 0.5 for elastic net with 0.5 alpha

# Standardize
p <- ncol(X)
n <- nrow(X)
xmn <- scale(X, center = FALSE, scale = rep(sqrt(n - 1), p))

fit <- glmnet(
  x = X,
  y = y,
  lambda = lambdas.rr,
  alpha = 0
  )

### Legend function - do copy and paste
lbs_fun <- function(fit, ...) {
  L <- length(fit$lambda)
  x <- log(fit$lambda[L])
  y <- fit$beta[, L]
  labs <- names(y)
  legend(
    'bottomleft',
    legend = labs,
    col = 1:length(labs),
    lty = 1,
    cex = 0.7
    ) 
}

plot(
  fit,
  xvar = "lambda",
  col = 1:dim(coef(fit))[1],
  main = "White Wine Data",
  lwd = 2
  )
lbs_fun(fit)
```

```r
## Cross validation
library(glmnet)
lambdas.rr <- exp(seq(-4, 4, length=50))
k <- 10
rr.cv <- cv.glmnet(x=X, y=y, lambda=lambdas.rr, nfolds=k, alpha=0)
fit <- glmnet(x=X, y=y, lambda=rr.cv$lambda.min, alpha=0)



plot(rr.cv)
```

- Compare the out of sample prediction of the cross-validated ridge regression with the linear model that only uses first three and ten covariates

### **Correct** way to cross validate

```r
set.seed(123)

  n <- length(y)
  k <- 10

  ii <- sample(rep(1:k, length= n))
  
  pr.1 <- pr.2 <- pr.l <- rep(NA, length(y))
  for (j in 1:k){
    hold <- (ii == j)
    train <- (ii != j)
    xx.tr <- xmn[train,]
    y.tr <-  y[train]
    xx.te <- xmn[hold,]
    
  coef_reg1 <- lm(y.tr ~ xx.tr[, c(1,2,3)])$coef
      
  coef_reg2 <- lm(y.tr ~ xx.tr[, c(1:10)])$coef
 
    ### find the best ridge on the TRAINING SET!
    rr.cv <- cv.glmnet(x=as.matrix(xx.tr), y=y.tr, lambda=lambdas.rr, nfolds=k, alpha=0)



    pr.1[hold] <- cbind(1, xx.te[, c(1,2,3)])%*%coef_reg1
    pr.2[hold] <- cbind(1, xx.te[, c(1:10)])%*%coef_reg2
    
    pr.l[hold] <- predict(rr.cv, newx= as.matrix(xx.te))
  }

mspe1 <- mean((pr.1-y)^2)
mspe2 <- mean((pr.2-y)^2)
mspe.ridge <- mean((pr.l-y)^2)
print(cbind(mspe1, mspe2, mspe.ridge))
## Note: this can be different if we did not cross-validated twice! 
```

## Bayesian Modeling: A brief Intro


Model: 
  $$
  Y_i = X_i \beta + \varepsilon_i, \quad \varepsilon_i \sim \mathcal{N}(0,1)
  $$
**Prior**:
$$
\beta  \sim \mathcal{N}(\mathbf{0}, \lambda^{-1} I_p)
$$
  
 

**Posterior Distribution**: use Bayes theorem
  
  $$
P(\beta | Y_1,..., Y_n, X_1, ..., X_n) = \frac{P(Y_1, ..., Y_n, X_1, ..., X_n | \beta)P(\beta)}{P(Y_1, ..., Y_n, X_1,..., X_n)}
  $$
After doing the math  we obtain 
 $$
 \beta | Y_1,..., Y_n, X_1, ..., X_n \sim \mathcal{N}(\mu_{bayes}, Q^{-1})
 $$
where 
$$   
\mu_{bayes} = (X^\top X + \lambda I_p)^{-1}X^\top Y, \quad Q = \lambda (I + X^\top X). 
$$
**Example:**
We want to predict the waiting time for a bus in Rome at a certain stop. To do so, we have collected data from Atac, the firm who manages the public traffic service between April and May 2016. We have at disposal 193 variables and the dependent variable is travel time between two points of the city. The only objective is to have an accurate out-of-sample predictive power. 
      
- Which model would you use? How would you choose the model?
        
- Open the text file Atacfinal and set the dependent variable the 160th columns 
      



```r
## Generate data 
data <- read.table("./data/Atacfinal.txt")

y <- data[,160]
xm <- data[,-160]
xm <- as.matrix(xm)
p <- dim(xm)[2]
## Start with a gaussian model for the outcome (is this a correct choice?)

## Set lambda = 1
lambda = 1
## posterior parameters: 
Q <- (t(xm)%*%xm + diag(lambda, dim(xm)[2]))/lambda
w_bayes <- solve(t(xm)%*%xm + diag(lambda, dim(xm)[2]), t(xm)%*%y)

## Draw parameters from the posterior: 
num_draws <- 1000
## Draw iid gaussian random variables 
random_draws <- matrix(rnorm(num_draws*p), nrow = p, ncol = num_draws)
## Rescale by the covariance matrix and add the mean
sampled_beta <- chol(solve(Q))%*%random_draws 
sampled_beta <- apply(sampled_beta, 2, function(x) x + w_bayes)

## Observe distribution of coefficients

## We look at the coefficient which correspond to the prediction of the bus travel agency

hist(sampled_beta[192,], main = 'Prediction 1 of the Company - Distribution of the coefficient')
```

## Bayesian Inference in R

- Very powerful program called Stan : https://mc-stan.org/users/interfaces/rstan
- Simple language with powerful sampling schemes (Hamiltonian Monte Carlo)



```r
# Generate data
N <- 300
p <- 30
beta <- c(1,1, rep(0, p-2))
X <- matrix(rnorm(p*N), nrow = N, ncol = p)
y <- X%*%beta + rnorm(N)
y <- as.numeric(y)
```

### Example 1: Linear Regression

$$
Y_i = X_i \beta + \varepsilon, \quad \varepsilon_i \sim \mathcal{N}(0, \sigma^2)
$$
(no prior specification)

### Stan Code


data {

    int<lower=2> N;
    int<lower=1> p;
    vector[N] y;
    matrix[N,p] X;
}


parameters {

    real<lower=0> sig_e;
    real beta0;
    vector[p] beta;
}


model {

    y ~ normal(rep_vector(beta0, N) + X * beta, sig_e);
}

```r
library(rstan)
rstan_options(auto_write = TRUE)
options(mc.cores = parallel::detectCores())

## Prepare the data 
data <-list(N = N, p = p, X = X, y = y)


## fit unpenalized model
model_basic <- stan("./additional_programs/model_basic.stan", data=data)
```

```r
plot(model_basic, pars="beta")
plot(model_basic, pars="sig_e")
```

## Ridge: stan code

data {

    int<lower=2> N;
    int<lower=1> p;
    vector[N] y;
    matrix[N,p] X;
}


parameters {

    real<lower=0> sig_e;
    real beta0;
    vector[p] beta;
    real<lower=0> lam;
}


model {

    beta ~ normal(0, lam * sig_e);
    y ~ normal(rep_vector(beta0, N) + X * beta, sig_e);
}

```r

## fit ridge model

model_ridge <- stan("./additional_programs/model_ridge.stan", data=data)
```

```r
plot(model_ridge, pars="beta")
plot(model_ridge, pars="sig_e")
plot(model_ridge, pars="lam")
```
